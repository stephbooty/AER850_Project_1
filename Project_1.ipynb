```python
#Step 4: Classification Model Development/Engineering - 20 Marks
from sklearn.model_selection import train_test_split

# Define the imput (X, Y, Z) and the predict variable (Step)
X_features = data[['X', 'Y', 'Z']]
y_target = data['Step']

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, test_size=0.2, random_state=42)

# Print the shapes of the splits to make sure it works
print("Training data:", X_train.shape)
print("Testing daya:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

#  Define the parameter grid for GridSearchCV
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs'],
    'penalty': ['l1', 'l2']
}

#  Initialize the Logistic Regression model
logistic_model = LogisticRegression(max_iter=1000)

#  Initialize GridSearchCV with cross-validation = 3
grid_search = GridSearchCV(estimator=logistic_model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

#  Fit GridSearchCV on the training data
grid_search.fit(X_train, y_train)

#  Print the best hyperparameters
print("Best Hyperparameters from GridSearchCV:", grid_search.best_params_)

#  Predict using the best model found by GridSearchCV
y_pred_grid_search = grid_search.best_estimator_.predict(X_test)

#  Evaluate the model's performance using accuracy
accuracy_grid_search = accuracy_score(y_test, y_pred_grid_search)
print("Logistic Regression with GridSearchCV Accuracy:", accuracy_grid_search)

#  Get the predicted probabilities for the test set
y_prob_grid_search = grid_search.best_estimator_.predict_proba(X_test)

#  Extract the probabilities for the correct class
y_prob_correct_grid_search = y_prob_grid_search[range(len(y_test)), y_test - 1]

#  Plot the distribution of predicted probabilities
plt.figure(figsize=(4, 4))
plt.hist(y_prob_correct_grid_search, bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Predicted Probabilities for Correct Class (Logistic Regression - GridSearchCV)')
plt.xlabel('Predicted Probability')
plt.ylabel('Frequency')
plt.show()

```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

#
param_grid = {
    'n_estimators': [50, 100, 200, 300],
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

#  Initialize the Random Forest model
rf_model = RandomForestClassifier(random_state=42)

#  Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=rf_model, param_distributions=param_grid,
                                   n_iter=10, cv=3, random_state=42, n_jobs=-1)

#  Fit RandomizedSearchCV on the training data
random_search.fit(X_train, y_train)

#  Print the best hyperparameters
print("Best Hyperparameters:", random_search.best_params_)

#  Predict using the best found model
y_pred_random_search = random_search.best_estimator_.predict(X_test)

#  Evaluate the model's performance using accuracy
accuracy_random_search = accuracy_score(y_test, y_pred_random_search)
print("Random Forest with RandomizedSearchCV Accuracy:", accuracy_random_search)

#  Get the predicted probabilities for the test set
y_prob_random_search = random_search.best_estimator_.predict_proba(X_test)

#  Extract the probabilities for the correct class
y_prob_correct_random_search = y_prob_random_search[range(len(y_test)), y_test - 1]

# Plot the distribution of predicted probabilities for the correct class
plt.figure(figsize=(4, 4))
plt.hist(y_prob_correct_random_search, bins=20, color='purple', alpha=0.7)
plt.title('Distribution of Predicted Probabilities for Correct Class (RandomizedSearchCV Random Forest)')
plt.xlabel('Predicted Probability')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initialize the k-NN model
knn_model = KNeighborsClassifier(n_neighbors=5)

#  Train the model on the training data
knn_model.fit(X_train, y_train)

#  Make predictions on the test data
y_pred_knn = knn_model.predict(X_test)

#  Evaluate the model's accuracy
accuracy_knn = accuracy_score(y_test, y_pred_knn)

# Print the accuracy of the k-NN model
print("k-NN Model Accuracy:", accuracy_knn)

#  Get the predicted probabilities  using predict_proba
y_prob_knn = knn_model.predict_proba(X_test)

#  Extract the probabilities for the correct class
y_prob_correct_knn = y_prob_knn[range(len(y_test)), y_test - 1]

#  Plot the distribution of predicted probabilities
plt.figure(figsize=(4,4 ))
plt.hist(y_prob_correct_knn, bins=20, color='orange', alpha=0.7)
plt.title('Distribution of Predicted Probabilities for Correct Class (k-NN)')
plt.xlabel('Predicted Probability')
plt.ylabel('Frequency')
plt.show()
```

```python
"""
Explanation of Metrics: Accuracy, Precision, F1 Score

  Accuracy:
    What It Means: Accuracy measures the proportion of correct predictions
    out of the total number of predictions.

  Precision:
    What It Means: Precision measures the proportion of positive predictions
    (in this case, predictions that maintenance is needed) that were correct.

  F1 Score:
    What It Means: The F1 score is the harmonic mean of precision and recall.
    It balances both precision (how the positive predictions are correct) and
    recall (how the actual positives are caught by the model). Its especially
    useful when you need to a balance false positives and false negatives.

Which Metric to Prioritize
  For this project, the choice depends on how costly or risky false positives
  false negatives are.

  F1 score will be prioritized. It gives a good balance between catching all
  necessary maintenance steps while also minimizing unnecessary maintenance.

The best model seemed was Knn we will use a confusion matrix for this.
"""
```

```python
#Step 6: Stacked Model Performance Analysis - 5 Marks
# Stacking Classifier Logistic Regression and Random Forest as base models, with a Logistic Regression learner

from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

#  Define the base models (Logistic Regression and Random Forest)
base_models = [
    ('logistic', LogisticRegression(max_iter=1000)),
    ('random_forest', RandomForestClassifier(n_estimators=50))
]

#  Define the meta-learner (we'll use Logistic Regression as well)
meta_learner = LogisticRegression(max_iter=1000)

#  Define the stacking classifier
stacking_model = StackingClassifier(estimators=base_models, final_estimator=meta_learner)

# S Train the stacking model on the training data
stacking_model.fit(X_train, y_train)

#  : Make predictions on the test data
y_pred_stacking = stacking_model.predict(X_test)

#  Function to calculate and print model performance metrics
def evaluate_model(y_test, y_pred, model_name):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted')

    print(f"Model: {model_name}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print("-" * 40)

#  Evaluate the performance of the stacked model
evaluate_model(y_test, y_pred_stacking, "Stacked Model")

#  Generate the confusion matrix for the stacked model
cm_stacking = confusion_matrix(y_test, y_pred_stacking)

#  Display the confusion matrix
disp_stacking = ConfusionMatrixDisplay(confusion_matrix=cm_stacking)
disp_stacking.plot(cmap='Blues')
plt.title('Confusion Matrix - Stacked Model')
plt.show()
```

```python
"""
Stacked Model Performance

  Performance Metrics:
    Accuracy: 97.09%
    Precision: 97.38%
    F1 Score: 97.05%

These are strong results, showing that the stacking classifier is performing
very well, with high accuracy and balanced precision and F1 scores.

Confusion Matrix Analysis:
  The confusion matrix shows that most predictions are correctly classified

  The highest values (Steps 7, 8, and 9) are classified correctly the most
  often, which aligns with the strong performance of the model.

  There are very few misclassifications, showing that the model is handling the
  data well.

The stacked models accuracy (97.09%) is slightly better than the individual
models (Logistic Regression, Random Forest, k-NN), showing that combining their
strengths leads to better overall performance.
"""
```
